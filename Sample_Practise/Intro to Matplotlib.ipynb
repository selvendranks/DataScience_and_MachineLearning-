# Intro to matplotlib

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y);
```

# matplotlib example workflow
WE want:

* Line plot
* Scatter plot
* Bar plot
* Histograms
* sub plots

```python
#figure size
fig ,ax = plt.subplots(figsize=(3,3))
ax.plot(x,y);

#customize plot
ax.set(title='Sample plot', xlabel='x-axis', ylabel='y-axis');

#save and show 
fig.savefig('sample-plot.png')
```

```python
#create some data
x = np.linspace(0,10,100)
#plot the data
fig, ax = plt.subplots();
ax.plot(x,x**2);
```

```python
#use same data to make a scatter
fig , ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#another example
fig , ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#create bar graph 
nut_butter_prices = {'Almond':10,
                     'peanut':8,
                      'casew':12}
fig,ax = plt.subplots();
ax.bar(nut_butter_prices.keys(),height=nut_butter_prices.values());
ax.set(title="Drag butter prices",xlabel='Products',ylabel='Price $');
```

```python
#create bar diagram horizontally , vlaues must be in list format
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
```

```python
# x represents the randn value and y represents the the count value of x
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

# subplots

```python
#subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
ax4.hist(x)
```

```python
#subplot option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
ax[1,1].hist(x)
```

```python
#removing symbols and last two zeroes 
car_sales["Price"] = car_sales['Price'].str.replace('[\$\.\,]','')
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
type(car_sales['Price'][0])
```

```python
#cumpulative sum of Prices
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#plot total sales pd pi
car_sales.plot(x='Sale Date',y='Total Sales')
```

```python
#using object oriented
fig,ax = plt.subplots(figsize=(15,5))
ax.plot(car_sales['Sale Date'],car_sales['Total Sales'])
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.head()
```

```python
heart_disease['age'].plot.hist(bins=30)
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
#finding patients with heart disease above age 50
over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
fig , ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=heart_disease['age'],
           y=heart_disease['chol'],
           c=heart_disease['target']);

#Customize
ax.set(title='Heart disease',
       xlabel='age',
       ylabel='cholestrol'
       )
#legends
ax.legend(*scatter.legend_elements(),title='Target');
ax.set_xlim([0,100])#sets x axis limit

#add horizontalline
ax.axhline(heart_disease['chol'].astype('int').mean(),linestyle='--');
ax.axvline(heart_disease['age'].astype('int').mean());
```

```python
#subplot of chol,age,thalach
fig , (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#x0
scatter = ax0.scatter(x=heart_disease['age'],
            y=heart_disease['chol'],
            c=heart_disease['target'])

ax0.set(
        title='Heart disease',
#        xlabel='age', 
       ylabel='cholestrol')

ax0.legend(*scatter.legend_elements(),title='Target');

ax0.axhline(heart_disease['chol'].astype('int').mean(),linestyle='--');

# ax1
ax1.scatter(x=heart_disease['age'],
            y=heart_disease['thalach'],
            c=heart_disease['target'])

ax1.set(
        title='Heart disease',
       xlabel='age',
       ylabel='Max heart rate'
       )

ax1.legend(*scatter.legend_elements(),title='Target');

ax1.axhline(heart_disease['thalach'].astype('int').mean(),linestyle='-.');

#add title to a fig

fig.suptitle('Heart Disease',fontsize=16,fontweight='bold');
```

```python
plt.style.use('seaborn-whitegrid')

fig , ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=heart_disease['age'],
           y=heart_disease['chol'],
           c=heart_disease['target'],
           cmap="winter" ); #color scheme

#Customize
ax.set(title='Heart disease',
       xlabel='age',
       ylabel='cholestrol'
       )
#legends
ax.legend(*scatter.legend_elements(),title='Target');
ax.set_xlim([0,100])#sets x axis limit

#add horizontalline
ax.axhline(heart_disease['chol'].astype('int').mean(),linestyle='--');
```

```python
plt.style.use('seaborn-whitegrid')

fig , (ax0,ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10),sharex=True)

#x0
scatter = ax0.scatter(x=heart_disease['age'],
            y=heart_disease['chol'],
            c=heart_disease['target'],
            cmap='winter')

ax0.set(
        title='Heart disease',
#        xlabel='age', 
       ylabel='cholestrol',
       )


ax0.legend(*scatter.legend_elements(),title='Target');

ax0.axhline(heart_disease['chol'].astype('int').mean(),linestyle='--');

# ax0.set_xlim([50,80]) #x axis is shared so no need in above plot
ax0.set_ylim([100,400])

# ax1
ax1.scatter(x=heart_disease['age'],
            y=heart_disease['thalach'],
            c=heart_disease['target'],
           cmap='winter',
            )

ax1.set(
        title='Heart disease',
       xlabel='age',
       ylabel='Max heart rate'
       )

ax1.set_xlabel("Age", color="red", fontdict={"size": 20})#customize labels

ax1.legend(*scatter.legend_elements(),title='Target');

ax1.axhline(heart_disease['thalach'].astype('int').mean(),linestyle='-.');

ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

#add title to a fig

fig.suptitle('Heart Disease',fontsize=16,fontweight='bold');
```
