# Intro to matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#method pyplot oriented
plt.plot();
```

```python
#y coordinates
plt.plot([1,2,3,9]);
```

```python
x = [1,2,3,10]
y = [5,6,7,1]
plt.plot(x,y);
```

```python
#1st method
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# 2nd method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y);
```

```python
#3rd recommended
fig,ax = plt.subplots()
ax.scatter(x,y),ax.plot(x,y);
```

# matplotlib example workflow
WE want:

* Line plot
* Scatter plot
* Bar plot
* Histograms
* sub plots

```python
#figure size
fig ,ax = plt.subplots(figsize=(3,3))
ax.plot(x,y);

#customize plot
ax.set(title='Sample plot', xlabel='x-axis', ylabel='y-axis');

#save and show 
fig.savefig('sample-plot.png')
```

```python
#create some data
x = np.linspace(0,10,100)
#plot the data
fig, ax = plt.subplots();
ax.plot(x,x**2);
```

```python
#use same data to make a scatter
fig , ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
#another example
fig , ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
#create bar graph 
nut_butter_prices = {'Almond':10,
                     'peanut':8,
                      'casew':12}
fig,ax = plt.subplots();
ax.bar(nut_butter_prices.keys(),height=nut_butter_prices.values());
ax.set(title="Drag butter prices",xlabel='Products',ylabel='Price $');
```

```python
#create bar diagram horizontally , vlaues must be in list format
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
```

```python
# x represents the randn value and y represents the the count value of x
x = np.random.randn(1000)
fig,ax = plt.subplots()
ax.hist(x);
```

# subplots

```python
#subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
ax4.hist(x)
```

```python
#subplot option 2
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

#plot to each axis
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10))
ax[1,0].barh(list(nut_butter_prices.keys()),width=list(nut_butter_prices.values()));
ax[1,1].hist(x)
```

# plotting from pandas dataframe

```python
import pandas as pd
```

```python
car_sales = pd.read_csv('car-sales.csv')
car_sales
```

```python
#plotting by panda series
ts = pd.Series([np.random.randint(0,20) for x in range(0,100)],index=range(0,100))
ts.plot()
```

```python
car_sales
```

```python
#removing symbols and last two zeroes 
car_sales["Price"] = car_sales['Price'].str.replace('[\$\.\,]','')
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range("1/1/2020",periods=len(car_sales))
car_sales
```

```python
type(car_sales['Price'][0])
```

```python
#cumpulative sum of Prices
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
#plot total sales pd pi
car_sales.plot(x='Sale Date',y='Total Sales')
```

```python
#using object oriented
fig,ax = plt.subplots(figsize=(15,5))
ax.plot(car_sales['Sale Date'],car_sales['Total Sales'])
```
