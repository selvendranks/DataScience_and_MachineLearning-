# manipulate data

```python
#convert make column to lower case
car_sales["Make"] = car_sales["Make"].str.lower()
car_sales
```

```python
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing
```

```python
# filling empty values with mean value of columns
for i in car_sales_missing.columns:
    if(car_sales_missing[i].dtype == float):
         car_sales_missing[i].fillna(car_sales_missing[i].mean(),inplace=True)  #inplace stores the value instead of we re-assigning
car_sales_missing
```

```python
# drops rows which has NaN values
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
car_sales_missing = pd.read_csv('car-sales-missing-data.csv')
car_sales_missing_dropped = car_sales_missing.dropna()
car_sales_missing_dropped.to_csv('car-sales-missing-data-dropped.csv')
```

```python
#creating new column and insert values
seats = pd.Series([5,5,5,5,5])
car_sales["Seats"] = seats
car_sales
```

```python
car_sales['Seats'].fillna(5,inplace=True)
car_sales
```

```python
# creating column from list (length of list must be same as number of rows , it wont consider NaN value like pd.Series())
fuel_economy = [2,1,4,3,6,7,3,6,8,4]
car_sales['Fuel per 100km'] = fuel_economy
car_sales
```

```python
car_sales["Fuel Used"] = car_sales['Odometer (KM)']/100 * car_sales['Fuel per 100km']
car_sales
```

```python
#create column by single value
car_sales['No of wheels'] = 4
car_sales
```

```python
#drop column
car_sales.drop("Seats",axis=1,inplace=True)
car_sales
```

```python

car_sales.style.hide_index()
```

```python
# shuffle rows frac determines fraction of rows 1 represent all rows 
car_sales_shuffled = car_sales.sample(frac=1)
car_sales_shuffled
```

```python
#reseting index 
car_sales_shuffled.reset_index(drop=True,inplace=True) # drop=True drops index of previous shuffled dataframe 
car_sales_shuffled
car_sales_shuffled
```

```python
#manipulating value with lambda function
car_sales['Odometer (KM)'] = car_sales['Odometer (KM)'].apply(lambda x: x/1.6)
car_sales
```
